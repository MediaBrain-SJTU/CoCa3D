```python
import numpy as np

object_centers = np.load('/GPFS/data/yhu/code_camera/OpenCOOD/object_centers.npy')[:,:4,:2].mean(axis=1)
agent_centers = np.load('/GPFS/data/yhu/code_camera/OpenCOOD/agent_centers.npy')
```

```python
agent_centers.shape
```

```python
from matplotlib import pyplot as plt 
plt.style.use('ggplot')

fontsize = 20
label_size = 16
# tick_size = 20
legend_size = 20
labelsize = 16
point_size = 100
figsize=(7,6)
figsize=(5.5,5.5)
params = {
    # 'legend.fontsize': 'x-large',
        # 'figure.figsize': (9, 7),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
plt.rcParams.update(params)

#######################  Performance v.s. Agent Number #######################
plt.tick_params(labelsize=labelsize)
figsize=(6.5,5)
step = 10
objects_dist, bins, _ = plt.hist(object_centers[:,0], bins = list(np.arange(-140, 150, step)), density=True, label='Objects') 
agents_dist, _, _ = plt.hist(agent_centers[:,0], bins = list(np.arange(-140, 150, step)), density=True, label='Agents') 
bins = bins + step/2
```

```python
figsize=(6.5,5)
fig = plt.figure(figsize=figsize)
ax = fig.add_subplot(111)

bins = bins + 140
plt.plot(bins[:-1], agents_dist, label='Agents', linewidth=3)
plt.plot(bins[:-1], objects_dist, label='Objects', linewidth=3)

ratio = 0.31
xleft, xright = ax.get_xlim()
ybottom, ytop = ax.get_ylim()
ax.set_aspect(abs((xright-xleft)/(ybottom-ytop))*ratio)

plt.xlabel('Field of view (0-280m)', size=label_size)
plt.ylabel('Object number', size=label_size)
plt.legend(loc='upper left', prop={'size': legend_size})
# plt.title("histogram") 
# plt.show()
plt.savefig('/GPFS/data/yhu/code_camera/OpenCOOD/opencood/visualization/abl_agent_dist/Object_distance_10agents.png')
```
